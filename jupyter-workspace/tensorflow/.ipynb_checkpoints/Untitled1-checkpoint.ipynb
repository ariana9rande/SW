```python
from tensorflow.keras import models, layers
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
from keras.utils import to_categorical
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('data/wine.csv', header = None)
df.head()
```

```python
df['quality'] = df[11]
df.head()
```

```python
del df[11]
```

```python
df.head()
```

```python
df['quality'].value_counts()
```

```python
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
y = y.values
print(type(y))
```

```python
y
```

```python
y = to_categorical(y, num_classes=11)
y.shape
```

```python
y[:3]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,
                                                    shuffle=True)
```

```python
min_max = MinMaxScaler()
X_scaled_train = min_max.fit_transform(X_train)
```

```python
X_scaled_train[:3]
```

```python
X_scaled_test = min_max.fit_transform(X_test)
```

```python
model = models.Sequential()
model.add(layers.Dense(30, input_dim=12, activation='relu'))
model.add(layers.Dense(24, activation='relu'))
model.add(layers.Dense(11, activation='softmax'))
model.summary()
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
early_stopping = EarlyStopping(monitor='val_loss', patience=20)
modelpath = 'model/wine_model.hdf5'
checkpoint = ModelCheckpoint(filepath=modelpath, monitor='val_loss', verbose=0, save_best_only= True)
```

```python
h = model.fit(X_scaled_train, y_train, epochs=30, batch_size=50,
             validation_split=0.25, verbose=1)
```

```python
score = model.evaluate(X_scaled_test, y_test)
print('Test Accuracy:%.4f' % (score[1]))
```
